# Exploratory Data Analysis - Titanic Dataset

This notebook performs exploratory data analysis on the Titanic dataset to understand the data distribution, missing values, and relationships between features.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
```

```python
# Load the Titanic dataset
df = sns.load_dataset('titanic')
print(f"Dataset shape: {df.shape}")
print(f"\nFirst 5 rows:")
df.head()
```

```python
# Basic information about the dataset
print("Dataset Info:")
df.info()

print("\nSummary Statistics:")
df.describe()
```

```python
# Check for missing values
print("Missing Values:")
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing_data, 'Missing %': missing_percent})
missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)
```

```python
# Target variable analysis
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='survived')
plt.title('Survival Distribution')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')

# Add percentage labels
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='survived', multiple='stack', bins=30)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Categorical features analysis
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Pclass vs Survived
sns.countplot(data=df, x='pclass', hue='survived', ax=axes[0,0])
axes[0,0].set_title('Passenger Class vs Survival')

# Sex vs Survived
sns.countplot(data=df, x='sex', hue='survived', ax=axes[0,1])
axes[0,1].set_title('Sex vs Survival')

# Embarked vs Survived
sns.countplot(data=df, x='embarked', hue='survived', ax=axes[1,0])
axes[1,0].set_title('Embarkation Port vs Survival')

# SibSp vs Survived
sns.countplot(data=df, x='sibsp', hue='survived', ax=axes[1,1])
axes[1,1].set_title('Siblings/Spouses vs Survival')

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap for numerical features
numerical_cols = ['age', 'fare', 'sibsp', 'parch', 'survived']
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix - Numerical Features')
plt.show()
```

```python
# Fare distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='pclass', y='fare', hue='survived')
plt.title('Fare Distribution by Class and Survival')
plt.yscale('log')
plt.show()
```

## Key Insights for Pipeline Development

1. **Missing Values**: Age (19.9%), Embarked (0.2%), Deck (77.2%)
2. **Target Imbalance**: ~38% survival rate
3. **Important Features**:
   - Sex: Females have much higher survival rate
   - Pclass: Higher classes have better survival
   - Age: Children have higher survival rates
   - Fare: Correlated with class and survival

### Preprocessing Strategy:
- **Numerical**: age, fare, sibsp, parch → imputation + scaling
- **Categorical**: sex, embarked, pclass → encoding
- **Drop**: deck (too many missing), who, adult_male, embark_town, alive, alone (redundant)
- **Target**: survived
